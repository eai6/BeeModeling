```python
joined_gdf = gpd.sjoin(gdff, pennsylvania_data, how="inner", op="within")
joined_gdf
```

```python
joined_gdf.plot(ax=plt.gca(), marker='o', color='red', markersize=5)
```

```python
joined_gdf.groupby('observed_on').count()
```

```python
def getYear(dt):
    try:
        return dt.split('-')[0]
    except:
        return np.nan

def getMonth(dt):
    try:
        return dt.split('-')[1]
    except:
        return np.nan
```

```python
joined_gdf['year'] = joined_gdf['observed_on'].apply(getYear)
joined_gdf['month'] = joined_gdf['observed_on'].apply(getMonth)
```

```python
joined_gdf.groupby(['year', 'month']).count()['id']#.plot(kind='bar', color='blue', figsize=(15,5))
```

```python
joined_gdf.groupby(['year']).count()['id'].plot(kind='bar', color='blue', figsize=(15,5))
```

```python
pennsylvania_data.plot(figsize=(20, 15))
#joined_gdf_2023 = joined_gdf[joined_gdf['year'] == '2023']
joined_gdf.plot(ax=plt.gca(), marker='o', color='red', markersize=5)
plt.show()
```

```python
pennsylvania_data.plot(figsize=(20, 15))
joined_gdf.plot(ax=plt.gca(), marker='o', c=joined_gdf['month'], cmap='plasma', markersize=5)
plt.show()
```
